```python
!pip install openpyxl
!pip install matplotlib
!pip install seaborn

import pandas as pd
# Load the Excel file
file_path = 'C:\\Users\\pc\\Documents\\Project_data analytics\\New folder\\excel\\international ed costs.xlsx'  
df = pd.read_excel(file_path, sheet_name='international ed costs')

# Preview the data
print(df.head())

import matplotlib.pyplot as plt

# Assuming you've already loaded your DataFrame as df
cost_components = ['Tuition_USD', 'Rent_USD', 'Visa_Fee_USD', 'Insurance_USD']
print(cost_components)
avg_components_by_country = df.groupby("Country")[cost_components].mean()

# Plotting
avg_components_by_country.plot(
    kind='barh',
    stacked=True,
    figsize=(12, 8),
    colormap='Set2'
)

plt.title("Average Education Cost Components by Country")
plt.xlabel("Average Cost (USD)")
plt.ylabel("Country")
plt.legend(title="Cost Component", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

for i in range(1,907) :
    if df.Tuition_USD[i] > 10000000:
        avg_components_by_country = df.groupby("Country")[cost_components].mean()

print(avg_components_by_country)


# Assuming you've already loaded your DataFrame as df
cost_components = ['Tuition_USD', 'Rent_USD', 'Visa_Fee_USD', 'Insurance_USD']
for i in range(1,907) :
    if df.Tuition_USD[i] > 500000:
        avg_components_by_country = df.groupby("Country")[cost_components].mean()
print(avg_components_by_country.head())
# Plotting
avg_components_by_country.head().plot(
    kind='barh',
    stacked=True,
    figsize=(12, 8),
    colormap='Set2'
)

plt.title("Average Education Cost Components by Country")
plt.xlabel("Average Cost (USD)")
plt.ylabel("Country")
plt.legend(title="Cost Component", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data (if not already loaded)
file_path = 'C:/Users/pc/Documents/Project_data analytics/New folder/excel/international ed costs.xlsx'
df = pd.read_excel(file_path, sheet_name='international ed costs')

# Set seaborn style
sns.set(style="whitegrid")

# Average Total Cost by Country
avg_cost_by_country = df.groupby("Country")["Total_Cost"].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_cost_by_country.values, y=avg_cost_by_country.index, palette="coolwarm")
plt.title("Average Total Education Cost by Country")
plt.xlabel("Average Total Cost (USD)")
plt.ylabel("Country")
plt.tight_layout()
plt.show()


top_universities = df[['University', 'Total_Cost']].sort_values(by='Total_Cost', ascending=False).head(5)

plt.figure(figsize=(10, 5))
sns.barplot(x='Total_Cost', y='University', data=top_universities, palette='Reds_r')
plt.title("Top 5 Most Expensive Universities")
plt.xlabel("Total Cost (USD)")
plt.ylabel("University")
plt.tight_layout()
plt.show()
```
